### Numpy 집계 함수(Aggregate Functions)
Numpy의 집계함수는 axis 인자를 기준으로 계산된다.
 - `axis=None` : 전체 행렬을 하나로 보고 계산
 - `axis=0` : row(x축)를 기준으로 각 row의 동일 인덱스를 묶어서 계산 -> row 요소가 사라짐
 - `axis=1` : col(y축)을 기준으로 각 column을 묶어서 계산 -> col 요소가 사라짐
 - `axis=2` : depth(z축) 를 기준으로 z축의 동일 인덱스 원소를 묶어서 계산 -> depth 요소가 사라짐

### Python array vs Numpy ndarray
Numpy의 경우 Vectorization을 통해 내부적으로 빠르게 행렬(배열) 연산을 처리한다.

`Numpy.ndarray`가 빠른이유

1. 대부분의 logic을 C 언어로 직접 구현
2. Python array와는 다르게 ndarray는 homogeneous data type
3. locality of reference
4. Single Instruction Multiple Data


- Python List
    - Heterogeneous type list
    - Linked list로 구현
    - 메모리 용량이 크고, dynamic type checking 때문에 속도가 느림
    
- Numpy.ndarray
    - Homogeneous type list
    - continuous memory layout
    - Cache 등에 있어서 Python List 보다 최적화됨,

```python
arr = range(1000000)
%time arr_2 = [i**2 for i in arr]
```

```python
n_arr = np.arange(1000000)
%time n_arr2 = n_arr**2
```
